```python
import os
import re
import pandas as pd
import pdfplumber

# Robust path handler
path_option_1 = "data/raw/igr/"
path_option_2 = "../data/raw/igr/"
folder_path = path_option_1 if os.path.exists(path_option_1) else path_option_2

file_name = "IGR_Tables_2014_complete.pdf"
full_path = os.path.join(folder_path, file_name)

extracted_13_14_rows = []

if os.path.exists(full_path):
    print(f"Extracting 2013 and 2014 timelines from Page 3 of: {file_name}...\n")
    with pdfplumber.open(full_path) as pdf:
        # Target Page 3 (Index 2)
        page = pdf.pages[2]
        lines = page.extract_text().split("\n")
        
        for line in lines:
            clean_line = line.strip()
            
            # Match pattern: (S/N digits)(Space)(State Name Letters)(Remaining numeric metrics)
            match = re.match(r"^(\d+)\s+([A-Za-z\s]+?)(?=\d)(.*)", clean_line)
            
            if match:
                sn = match.group(1)
                state = match.group(2).strip()
                remainder = match.group(3).strip()
                
                # Find all well-formed financial strings
                financial_numbers = re.findall(r"[\d,.]+", remainder)
                
                # According to the 5-year timeline structure (2010, 2011, 2012, 2013, 2014):
                # Index 3 is 2013 IGR, Index 4 is 2014 IGR
                val_2013 = financial_numbers[3] if len(financial_numbers) > 3 else None
                val_2014 = financial_numbers[4] if len(financial_numbers) > 4 else None
                
                extracted_13_14_rows.append({
                    "S/N": sn,
                    "State": state,
                    "2013_IGR": val_2013,
                    "2014_IGR": val_2014
                })

    # Convert to DataFrame grid
    df_13_14 = pd.DataFrame(extracted_13_14_rows)
    
    # --- Data Quality Audit Block ---
    print("--- Data Quality Audit: Checking for Blanks/Missing Values ---")
    missing_check = df_13_14[df_13_14.isnull().any(axis=1)]
    if missing_check.empty:
        print("✅ Clean Data Sweep! Both 2013 and 2014 are fully loaded for all states with zero missing values.")
    else:
        print(f"⚠️ Found missing metrics in these records:\n", missing_check)
    print("-" * 75)
    
    print("\n--- Displaying 2013–2014 Timeline Preview ---")
    print(df_13_14.head(15))
else:
    print(f"Error: Missing file {file_name} at path {folder_path}")
```

```python
import os
import re
import pandas as pd
import pdfplumber

# Robust path handler
path_option_1 = "data/raw/igr/"
path_option_2 = "../data/raw/igr/"
folder_path = path_option_1 if os.path.exists(path_option_1) else path_option_2

file_name = "IGR_Tables_2014_complete.pdf"
full_path = os.path.join(folder_path, file_name)

all_13_14_rows = []

if os.path.exists(full_path):
    print(f"Extracting 2013/2014 timelines from Pages 3 AND 4 of: {file_name}...\n")
    with pdfplumber.open(full_path) as pdf:
        
        # Loop through Page 3 (Index 2) and Page 4 (Index 3) to get all states
        for page_num in [2, 3]:
            if page_num < len(pdf.pages):
                page = pdf.pages[page_num]
                lines = page.extract_text().split("\n")
                
                for line in lines:
                    clean_line = line.strip()
                    
                    # Match pattern: (S/N digits)(Space)(State Name Letters)(Remaining metrics)
                    match = re.match(r"^(\d+)\s+([A-Za-z\s]+?)(?=\d)(.*)", clean_line)
                    
                    if match:
                        sn = match.group(1)
                        state = match.group(2).strip()
                        remainder = match.group(3).strip()
                        
                        financial_numbers = re.findall(r"[\d,.]+", remainder)
                        
                        # Dynamic position index verification for 5-year table layout
                        # Index 3 = 2013 IGR, Index 4 = 2014 IGR
                        val_2013 = financial_numbers[3] if len(financial_numbers) > 3 else None
                        val_2014 = financial_numbers[4] if len(financial_numbers) > 4 else None
                        
                        all_13_14_rows.append({
                            "S/N": int(sn),
                            "State": state,
                            "2013_IGR": val_2013,
                            "2014_IGR": val_2014
                        })

    # Convert to DataFrame grid and sort by S/N to keep it perfectly chronological
    df_13_14_full = pd.DataFrame(all_13_14_rows).sort_values(by="S/N").reset_index(drop=True)
    
    # --- Data Quality Audit Block ---
    print("--- 1. Data Quality Audit: Checking Complete State Coverage ---")
    print(f"Total Unique Rows Captured: {len(df_13_14_full)} out of 37 expected (36 States + FCT summary components)")
    
    missing_check = df_13_14_full[df_13_14_full.isnull().any(axis=1)]
    if missing_check.empty and len(df_13_14_full) >= 36:
        print("✅ Clean Data Sweep! Every single state across both pages is fully recovered with zero NaNs.")
    else:
        print(f"⚠️ Warning: Missing metrics or incomplete state rows found:\n", missing_check)
    print("-" * 75)
    
    print("\n--- 2. Complete 2013–2014 Timeline Preview ---")
    print(df_13_14_full)
else:
    print(f"Error: Missing file {file_name}")
```

```python
import pdfplumber
import re
import pandas as pd

path = "data/raw/igr/"
file_name = [f for f in os.listdir(path) if "2016" in f][0]
full_path = os.path.join(path, file_name)

extracted_data = []

with pdfplumber.open(full_path) as pdf:
    # Iterate through all pages
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            # 1. Identify the State Name (usually at the top of the page)
            state_match = re.search(r"([A-Za-z\s]+)\s+State", text)
            if state_match:
                state_name = state_match.group(1).strip()
                
                # 2. Extract the 2016 value specifically
                # We look for '2016' followed by a number with commas and decimals
                val_match = re.search(r"2016\s+([\d,]+\.\d+)", text)
                
                if val_match:
                    value_2016 = val_match.group(1)
                    extracted_data.append({"State": state_name, "2016_IGR": value_2016})
                    print(f"Found: {state_name} -> {value_2016}")

df_2016 = pd.DataFrame(extracted_data)
print(df_2016.head())
```
